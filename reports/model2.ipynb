# Logistic regression in Tensorflow

The logistic regression model was done using a multiplicative layer and then passed thorugh a softmax.
The optimizer was gradient descent. The input for the model was a resized image 16x16.

Some parameters of the model:

- First layer size: [16*16*3,43]
- Output size: [43] (One hot encoding for the classes)
- Trainded with Gradient Descent
- Loss function: cross entropy
- Epochs for training: 500
- Learning rate: 0.1

Additional preprocessing:

- Images where resized to [16,16,3] and then converted to one vector of size[16,16,3]
- Pixels were standarized using scikit learn function
- Output class was transformed using one-hot enconding
